# Chapter 6 - Selecting Subsets of Data

### Selecting Series Data

```python
import pandas as pd
import numpy as np

college = pd.read_csv("https://raw.githubusercontent.com/PacktPublishing/Pandas-Cookbook-Second-Edition/master/data/college.csv", index_col="INSTNM")
```

```python
city = college["CITY"]
city
```

```python
# pull scalar directly from Series
city['Alabama A & M University']
```

```python
# pull scalar value using .loc by name
city.loc['Alabama A & M University']
```

```python
# pull scalar using .iloc by position
city.iloc[0]
```

```python
# pull multiple values by indexing
# passing a list returns a Series instead of a scalar
city[['Alabama A & M University', 'Alabama State University']]
```

```python
# repeat using .loc
city.loc[['Alabama A & M University', 'Alabama State University']]
```

```python
# repeat using .iloc
city.iloc[[0, 4]]
```

```python
# use a slice to pull multiple values
city['Alabama A & M University':'Alabama State University']
```

```python
# use a slice to pull multiple values by position
city[0:5]
```

```python
# use a slice to pull multiple values using .loc
city.loc['Alabama A & M University':'Alabama State University']
```

```python
# use a slice to pull multiple values using .iloc
city.iloc[0:5]
```

```python
# use Boolean to pull certain values
alabama_mask = city.isin(["Birmingham", "Montgomery"])
city[alabama_mask]
```

### Selecting DataFrame rows

```python
# most explicit & preferred way to select df rows is with .loc or .iloc
# both are capable of selecting by rows or by rows & cols
college.sample(5, random_state=42)
```

```python
# select entire row at position
college.iloc[60]
```

```python
# same as prev using .loc
college.loc["University of Alaska Anchorage"]
```

```python
# select disjointed set of rows as a df
college.iloc[[60, 99, 3]]
```

```python
# same as prev using .loc
labels = ["University of Alaska Anchorage", "International Academy of Hair Design",
          "University of Alabama in Huntsville"]

college.loc[labels]
```

```python
# use slice notation w/ .iloc to select contiguous rows
college.iloc[99:102]
```

```python
# slice notation w/ .loc is a closed interval (includes both start & stop label)
start = "International Academy of Hair Design"
stop = "Mesa Community College"
college.loc[start:stop]
```

```python
# the list of index labels can be selected directly from the df
college.iloc[[60, 99, 3]].index.tolist()
```

### Selecting DataFrame rows & columns simultaneously
image.png

```python
# select first three rows and first four columns w/ slice notation
college.iloc[:3, :4]
```

```python
# select all rows of diff cols
college.iloc[:, [4, 6]].head()
```

```python
college.loc[:, ["WOMENONLY", "SATVRMID"]].head()
```

```python
# select disjointed rows & cols
college.iloc[[100, 200], [7, 15]]
```

```python
# select single scalar value
college.iloc[5, -4]
```

```python
college.loc["The University of Alabama", "PCTFLOAN"]
```

```python
# slice the rows & select a single column
college.iloc[90:80:-2, 5]
```

```python
start = "Empire Beauty School-Flagstaff"
stop = "Arizona State University-Tempe"
college.loc[start:stop:-2, "RELAFFIL"]
```

```python
# to select only rows (along w/ all cols), it is not necessary to use a colon following a comma
# the default is to select all cols if no comma is present
college.iloc[:10]
```

```python
# equivalent to prev
college.iloc[:10, :]
```

### Selecing data with both integers and labels

```python
# use index method .get_loc to find integer position of desired col
col_start = college.columns.get_loc("UGDS_WHITE")
col_end = college.columns.get_loc("UGDS_UNKN") + 1
col_start, col_end
```

```python
# use col_start & col_end to select cols by position using .iloc
college.iloc[:5, col_start:col_end]
```

```python
# select the 10th through 15th (inclusive) rows along w/ cols UGDS_WHITE and UGDS_UNKN
row_start = college.index[10]
row_end = college.index[15]
college.loc[row_start:row_end, "UGDS_WHITE":"UGDS_UNKN"]
```

```python
# can replicate prev by chaining .loc & .iloc
college.iloc[10:16].loc[:, "UGDS_WHITE":"UGDS_UNKN"]
```

### Slicing lexicographically

```python
# only works if the index is sorted
# select all college w/ names b/t Sp and Su
# attempt w/o sorting

college.loc["Sp":"Su"]
```

```python
# sort the index & re-attempt prev
college = college.sort_index()
college.loc["Sp":"Su"]
```
