```python
# change "city08" & "comb08" columns to type int16 in order to save ~75% of memory usage
fueleco[["city08", "comb08"]].info(memory_usage="deep")
```

```python
fueleco[["city08", "comb08"]].assign(
    city08=fueleco.city08.astype(np.int16),
    comb08=fueleco.comb08.astype(np.int16)).info(memory_usage="deep")
```

Note there is an analogous finfo function in NumPy for retrieving float info.
An option for conserving memory for string cols is to convert to categories if each value for a string col is unique, this will slow down pandas and use more memory. If you have low cardinality, you can save a lot of memory

### Categorical Data

We can broadly classify data in to:


1.   dates
2.   continuous values
3.   categorical values

In this section, we will explore quantifying & visualizing categorical data.

```python
# find object data type columns
fueleco.select_dtypes(object).columns
```

```python
# use nunique to determine cardinality
fueleco.drive.nunique()
```

```python
# use sample to see some of the values
fueleco.drive.sample(5, random_state=42)
```

```python
# determine # and % of missing values
fueleco.drive.isna().sum()
```

```python
fueleco.drive.isna().mean() * 100
```

```python
# use value_counts to summarize a col
fueleco.drive.value_counts()
```

```python
# if too many values in summary above, look at top n and collapse the rest
top_n = fueleco.make.value_counts().index[:6]

fueleco.assign(make=fueleco.make.where(fueleco.make.isin(top_n), "Other")).make.value_counts()
```

```python
# use pandas to plot the counts & visualize
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(10, 8))

fueleco.assign(make=fueleco.make.where(fueleco.make.isin(top_n), "Other")).make.value_counts().plot.bar(ax=ax)
fig.savefig("c5-catpan.png", dpi=300)
```

```python
# use seaborn to plot counts & visualize
import seaborn as sns
sns.countplot(y="make", data=(fueleco.assign(make=fueleco.make.where(fueleco.make.isin(top_n), "Other"))))
fig.savefig("c5-catsns.png", dpi=300)
```

```python
# look at rows where drive is missing
fueleco[fueleco.drive.isna()]
```

```python
# value_counts does not show missing values by default. Can use dropna parameter to fix
fueleco.drive.value_counts(dropna=False)
```

```python
# some cols that report object data types aren't really categorical
fueleco.rangeA.value_counts(dropna=False)
```

The rangeA column is showing as an object b/c of slashes (/) and dashes (-) in some of the entries, causing Pandas to not know how to convert those values to a number.

```python
# use the str.extract method along w/ regex to find offending characters
fueleco.rangeA.str.extract(r"([^0-9.])").dropna().apply(lambda row: "".join(row),
                                                        axis=1).value_counts()
```

```python
# the rangeA col actually has two types: float & string
# reported as object b/c it can hold heterogenous typed cols
# missing values are stored as NaN and non-missing as strings
set(fueleco.rangeA.apply(type))
```

```python
# count of missing values
fueleco.rangeA.isna().sum()
```
