```python
# use nunique to determine cardinality
fueleco.drive.nunique()
```

```python
# use value_counts to summarize a col
fueleco.drive.value_counts()
```

```python
# if too many values in summary above, look at top n and collapse the rest
top_n = fueleco.make.value_counts().index[:6]

fueleco.assign(make=fueleco.make.where(fueleco.make.isin(top_n), "Other")).make.value_counts()
```

```python
# use pandas to plot the counts & visualize
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(10, 8))

fueleco.assign(make=fueleco.make.where(fueleco.make.isin(top_n), "Other")).make.value_counts().plot.bar(ax=ax)
fig.savefig("c5-catpan.png", dpi=300)
```

```python
# use seaborn to plot counts & visualize
import seaborn as sns
sns.countplot(y="make", data=(fueleco.assign(make=fueleco.make.where(fueleco.make.isin(top_n), "Other"))))
fig.savefig("c5-catsns.png", dpi=300)
```

```python
# look at rows where drive is missing
fueleco[fueleco.drive.isna()]
```

```python
# value_counts does not show missing values by default. Can use dropna parameter to fix
fueleco.drive.value_counts(dropna=False)
```

```python
# some cols that report object data types aren't really categorical
fueleco.rangeA.value_counts(dropna=False)
```

The rangeA column is showing as an object b/c of slashes (/) and dashes (-) in some of the entries, causing Pandas to not know how to convert those values to a number.

```python
# use the str.extract method along w/ regex to find offending characters
fueleco.rangeA.str.extract(r"([^0-9.])").dropna().apply(lambda row: "".join(row),
                                                        axis=1).value_counts()
```

```python
# the rangeA col actually has two types: float & string
# reported as object b/c it can hold heterogenous typed cols
# missing values are stored as NaN and non-missing as strings
set(fueleco.rangeA.apply(type))
```

```python
# count of missing values
fueleco.rangeA.isna().sum()
```

According to the data source, the 'rangeA' value represents the range for a 2nd fuel type on dual fuel vehicles. Using Pandas, we can replace the missing values with zero, replace dashes with slashes, then split and take the mean value for each row (in the case of a dash/slash).

```python
fueleco.rangeA.fillna("0").str.replace("-", "/").str.split("/", expand=True).astype(float).mean(axis=1)
```

```python
# could also treat numeric cols as categories by binning them
# using cut and qcut Pandas functions
# cut creates equal-width bins or bin widths we specify
# for the rangeA col, most values were empty (so replaced w/ zero)
# 10 equal width bins
fueleco.rangeA.fillna("0")\
  .str.replace("-", "/")\
  .str.split("/", expand=True)\
  .astype(float)\
  .mean(axis=1)\
  .pipe(lambda ser_: pd.cut(ser_, 10))\
  .value_counts()
```

```python
# qcut (quantile cut) will cut the entries into bins w/ the same size
# b/c rangeA is heavily skewed & most entries are zero, qcut will fail
# it does (somewhat) work w/ city08, where values are whole numbers
# so they don't bin evenly, but sizes are close
fueleco.rangeA.fillna("0")\
  .str.replace("-", "/")\
  .str.split("/", expand=True)\
  .astype(float)\
  .mean(axis=1)\
  .pipe(lambda ser_: pd.qcut(ser_, 10))\
  .value_counts()
```

```python
fueleco.city08.pipe(
    lambda ser: pd.qcut(ser, q=10)
).value_counts()
```

### Comparing continuous values across categories

Compare continuous variables in different categories. We will look at mileage numbers in different brands: Ford, Honda, Tesla & BMW.

```python
# make mask for brands we want to use
mask = fueleco.make.isin(["Ford", "Honda", "Tesla", "BMW"])
# use group by to look at mean & std dev
fueleco[mask].groupby("make").city08.agg(["mean", "std"])
```

```python
# visualize city08 values for each make
g = sns.catplot(x="make", y="city08", data=fueleco[mask], kind="box")
g.ax.figure.savefig("c5-catbox.png", dpi=300)
```

```python
fueleco[mask].groupby("make").city08.count()
```

```python
# swarm plot on top of box plot
g = sns.catplot(x="make", y="city08", data=fueleco[mask], kind="box")
sns.swarmplot(x="make", y="city08", data=fueleco[mask], color="k", size=1, ax=g.ax)
g.ax.figure.savefig("c5-catbox2.png", dpi=300)
```

```python
# catplot can show more than the two dimensions above (city mpg & make)
# facet the grid by another feature
g = sns.catplot(x="make", y="city08", data=fueleco[mask], kind="box", col="year",
                col_order=[2012,2014,2016,2018], col_wrap=2)

g.axes[0].figure.savefig("c5-catboxcol.png", dpi=300)
```
