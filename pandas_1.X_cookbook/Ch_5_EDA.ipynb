The rangeA column is showing as an object b/c of slashes (/) and dashes (-) in some of the entries, causing Pandas to not know how to convert those values to a number.

```python
# use the str.extract method along w/ regex to find offending characters
fueleco.rangeA.str.extract(r"([^0-9.])").dropna().apply(lambda row: "".join(row),
                                                        axis=1).value_counts()
```

```python
# the rangeA col actually has two types: float & string
# reported as object b/c it can hold heterogenous typed cols
# missing values are stored as NaN and non-missing as strings
set(fueleco.rangeA.apply(type))
```

```python
# count of missing values
fueleco.rangeA.isna().sum()
```

According to the data source, the 'rangeA' value represents the range for a 2nd fuel type on dual fuel vehicles. Using Pandas, we can replace the missing values with zero, replace dashes with slashes, then split and take the mean value for each row (in the case of a dash/slash).

```python
fueleco.rangeA.fillna("0").str.replace("-", "/").str.split("/", expand=True).astype(float).mean(axis=1)
```

```python
# could also treat numeric cols as categories by binning them
# using cut and qcut Pandas functions
# cut creates equal-width bins or bin widths we specify
# for the rangeA col, most values were empty (so replaced w/ zero)
# 10 equal width bins
fueleco.rangeA.fillna("0")\
  .str.replace("-", "/")\
  .str.split("/", expand=True)\
  .astype(float)\
  .mean(axis=1)\
  .pipe(lambda ser_: pd.cut(ser_, 10))\
  .value_counts()
```

```python
# qcut (quantile cut) will cut the entries into bins w/ the same size
# b/c rangeA is heavily skewed & most entries are zero, qcut will fail
# it does (somewhat) work w/ city08, where values are whole numbers
# so they don't bin evenly, but sizes are close
fueleco.rangeA.fillna("0")\
  .str.replace("-", "/")\
  .str.split("/", expand=True)\
  .astype(float)\
  .mean(axis=1)\
  .pipe(lambda ser_: pd.qcut(ser_, 10))\
  .value_counts()
```

```python
fueleco.city08.pipe(
    lambda ser: pd.qcut(ser, q=10)
).value_counts()
```

### Comparing categorical values with categorical values

```python
# lower the cardinality
# limit the VClass column to six values, in a simple class column: SClass
# only use Ford, Tesla, BMW & Toyota

def generalize(ser, match_name, default):
  seen = None
  for match, name in match_name:
    mask = ser.str.contains(match)
    if seen is None:
      seen = mask
    else:
      seen |= mask
    ser = ser.where(~mask, name)
  ser = ser.where(seen, default)
  return ser

makes = ["Ford", "Tesla", "BMW", "Toyota"]

data = fueleco[fueleco.make.isin(makes)].assign(SClass = lambda df_: generalize(
    df_.VClass,
    [
        ("Seaters", "Car"),
        ("Car", "Car"),
        ("Utility", "SUV"),
        ("Truck", "Truck"),
        ("Van", "Van"),
        ("van", "Van"),
        ("Wagon", "Wagon")
    ],
    "other"))
```

```python
# summarize the counts of vehicle classes for each make
data.groupby(["make", "SClass"]).size().unstack()
```

```python
# use crosstab
pd.crosstab(data.make, data.SClass)
```

```python
# add more dimensions
pd.crosstab(
    [data.year, data.make], [data.SClass, data.VClass]
)
```

```python
# use Cramer's V measure to indicate categorical correlation
def cramers_v(x, y):
  confusion_matrix = pd.crosstab(x, y)
  chi2 = ss.chi2_contingency(confusion_matrix)[0]
  n = confusion_matrix.sum().sum()
  phi2 = chi2 / n
  r, k = confusion_matrix.shape
  phi2corr = max(0, phi2 - ((k-1) * (r-1)) / (n-1))
  rcorr = r - ((r-1) ** 2) / (n-1)
  kcorr = k - ((k-1) ** 2) / (n-1)
  return np.sqrt(phi2corr / min((kcorr - 1), (rcorr - 1)))

cramers_v(data.make, data.SClass)
```

```python
# the .corr method accepts a callable
data.make.corr(data.SClass, cramers_v)
```

```python
# visualize crosstab
fig, ax = plt.subplots(figsize=(8, 8))

(
    data.pipe(lambda df_: pd.crosstab(df_.make, df_.SClass)).plot.bar(ax=ax)
)

fig.savefig("c5-bar.png", dpi=300, bbox_inches="tight")
```

```python
# viz using seaborn bar chart
res = sns.catplot(kind="count", x="make", hue="SClass", data=data)

res.fig.savefig("c5-barsns.png", dpi=300, bbox_inches="tight")
```

```python
# viz relative sizes of groups by normalizing cross tab & making stacked bar
fig, ax = plt.subplots(figsize=(8, 8))

(
    data.pipe(
        lambda df_: pd.crosstab(df_.make, df_.SClass)
    )
    .pipe(lambda df_: df_.div(df_.sum(axis=1), axis=0))
    .plot.bar(stacked=True, ax=ax)
)

fig.savefig("c5-barstacked.png", dpi=300, bbox_inches="tight")
```
