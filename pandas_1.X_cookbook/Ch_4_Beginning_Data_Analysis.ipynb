### Reducing memory usage by changing data types

```python
diff_cols = ["RELAFFIL", "SATMTMID", "CURROPER", "INSTNM", "STABBR"]

col2 = college.loc[:, diff_cols]

col2.head()
```

```python
col2.dtypes
```

```python
original_mem = col2.memory_usage(deep=True)
original_mem
```

```python
# no need to use 64 bits for RELAFFIL as it contains only 0 or 1
# convert to 8-bit (1 byte) integer using .astype method
col2["RELAFFIL"]  = col2["RELAFFIL"].astype(np.int8)
col2.dtypes
```

```python
col2.memory_usage(deep=True)
```

```python
# change object data types to categorical
# if they have reasonably low cardinality (unique values)

# check unique values for each object col
col2.select_dtypes(include=["object"]).nunique()
```

```python
# STABBR is the only good candidate
col2["STABBR"] = col2["STABBR"].astype("category")
col2.dtypes
```

```python
new_mem = col2.memory_usage(deep=True)
new_mem
```

```python
# compare original memory usage to updated memory usage
new_mem / original_mem
```

### Selecting the smallest of the largest

```python
movie = pd.read_csv('https://raw.githubusercontent.com/PacktPublishing/Pandas-Cookbook-Second-Edition/master/data/movie.csv')
# select columns
movie2 = movie[["movie_title", "imdb_score", "budget"]]
movie2.head()
```

```python
# select top 100 movies using .nlargest
movie2.nlargest(100, "imdb_score").head()
```

```python
# chain the .nsmallest method to return the 5 lowest budget films among top 100 imdb_score
movie2.nlargest(100, "imdb_score").nsmallest(5, "budget")
```

### Selecting the largest of each group by sorting

```python
movie3 = movie[["movie_title", "title_year", "imdb_score"]]
# use .sort_values to sort the df by title_year
movie3.sort_values("title_year", ascending=True).head()
```

```python
# sort by multiple columns using a list
movie3.sort_values(["title_year", "imdb_score"], ascending=False)
```

```python
# use .drop_duplicates to keep only the first row of every year
movie3.sort_values(["title_year", "imdb_score"], ascending=False).drop_duplicates(subset="title_year")
```

```python
# sort on column ascending while simultaneously sorting another column descending
movie4 = movie[["movie_title", "title_year", "content_rating", "budget"]]
movie4.sort_values(["title_year", "content_rating", "budget"], ascending=[False, False, True])\
                  .drop_duplicates(subset=["title_year", "content_rating"])
```

### Calculating a trailing stop order price

*stop order* - an order placed by an investor to buy or sell a stock that executes whenever the market price reaches a certain point. Useful to prevent huge losses and protect gains.

*trailing stop order* - continually modify the sale price of the stop order a set % below the current market value. Never moves down and is always tied to the maximum value since the time of purchase.
